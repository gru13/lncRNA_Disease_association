```python
import networkx as nx 
G = nx.MultiGraph()
nodes = []
for a in df:
    nodes.append(df[a].unique().tolist())
    G.add_nodes_from(nodes[-1].copy())
for a in df.values:
    G.add_edge(a[0],a[1])
    G.add_edge(a[1],a[2])
    G.add_edge(a[2],a[0])
```
